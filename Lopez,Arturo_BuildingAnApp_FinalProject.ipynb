```python
import tkinter as tk
from tkinter import ttk
import sqlite3
from tkinter import messagebox

# This will setup the database  
def setup_db():
    conn = sqlite3.connect('user_data.db')
    c = conn.cursor()
    c.execute('''
        CREATE TABLE IF NOT EXISTS users (
            id INTEGER PRIMARY KEY,
            name TEXT NOT NULL,
            age INTEGER NOT NULL
        )
    ''')
    conn.commit()
    conn.close()

# Defining the app
class App:
    def __init__(self, root):
        self.root = root
        self.root.title("User Data Entry")
        self.root.geometry("300x200")

        # Name Label and Entry
        self.name_label = tk.Label(root, text="Name:")
        self.name_label.grid(column=0, row=0, padx=10, pady=10)
        self.name_entry = tk.Entry(root)
        self.name_entry.grid(column=1, row=0, padx=10, pady=10)

        # Age Label and Entry
        self.age_label = tk.Label(root, text="Age:")
        self.age_label.grid(column=0, row=1, padx=10, pady=10)
        self.age_entry = tk.Entry(root)
        self.age_entry.grid(column=1, row=1, padx=10, pady=10)

        # Save Button
        self.save_button = tk.Button(root, text="Save", command=self.save_data)
        self.save_button.grid(column=1, row=2, padx=10, pady=10)

    def save_data(self):
        name = self.name_entry.get()
        age = self.age_entry.get()

        if name and age.isdigit():
            conn = sqlite3.connect('user_data.db')
            c = conn.cursor()
            c.execute('INSERT INTO users (name, age) VALUES (?, ?)', (name, int(age)))
            conn.commit()
            conn.close()
            messagebox.showinfo("Success", "Data saved successfully!")
            self.name_entry.delete(0, tk.END)
            self.age_entry.delete(0, tk.END)
        else:
            messagebox.showerror("Error", "Invalid input. Please enter a valid name and age.")

if __name__ == "__main__":
    setup_db()
    root = tk.Tk()
    app = App(root)
    root.mainloop()

setup_db()
```
